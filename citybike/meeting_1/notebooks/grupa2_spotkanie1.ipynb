# Z uwzględnieniem że trasa z A->B = B->A

```python
stacje = list(wyp_2018['departure'].unique())


def trasy(dep,ret):
    for x in range(0,len(wyp_2018)-1):
        if dep[x] != ret[x]:
            for word in stacje[1:]:
                if ret[x] in stacje[:stacje.index(word)]:
                    a = dep[x]
                    dep[x] = ret[x]
                    ret[x] = a

```

```python
dep = list(wyp_2018['departure'])
ret = list(wyp_2018['return'])

trasy(dep,ret)

wyp_2018['departure']= dep
wyp_2018['return'] = ret
```

```python
(wyp_2018['departure'] + " -- " + wyp_2018['return']).value_counts()
```

# 7) Dzienny rekord wypożyczeń
