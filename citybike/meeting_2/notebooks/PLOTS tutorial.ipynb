<h3> Let's start with MATPLOTLIB!

A Figure object is the outermost container for a matplotlib graphic, which can contain multiple Axes objects. One source of confusion is the name: an Axes actually translates into what we think of as an individual plot or graph (rather than the plural of “axis,” as we might expect). Below the Axes in the hierarchy are smaller objects such as tick marks, individual lines, legends, and text boxes. Almost every “element” of a chart is its own manipulable Python object.

The prescribed way to create a Figure with a single Axes is (not too intuitively) with plt.subplots(). The default call is subplots(nrows=1, ncols=1). Let's see:

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots()
```

Notice that we don’t pass arguments to subplots() here. Consequently, ax is a single AxesSubplot object:

```python
type(ax)
```

We can call instance methods to create the line plot:

```python
#plt.plot(x_data,y_data)
plt.plot([1,2,3,4],[2,4,6,8])
```

Or a scatter plot:

```python
#plt.scatter(x_data,y_data)
plt.scatter([1,2,3,4],[2,4,6,8])
```

Or a bar plot:

```python
plt.bar([1,2,3,4],[2,4,6,8])
```

You can also update the layout:

```python
fig, ax = plt.subplots()
plt.scatter([1,2,3,4],[2,4,6,8])
ax.set_title('This is a title')
ax.set_xlabel('this is an x axis label')
ax.set_ylabel('this is an y axis label')
```

<h2>PLOTLY

The plotly Python library (plotly.py) is an interactive, open-source plotting library that supports over 40 unique chart types covering a wide range of statistical, financial, geographic, scientific, and 3-dimensional use-cases.

Built on top of the Plotly JavaScript library (plotly.js), plotly.py enables Python users to create beautiful interactive web-based visualizations that can be displayed in Jupyter notebooks, saved to standalone HTML files, or served as part of pure Python-built web applications using Dash.

Plotly seems to be really simple:

```python
import plotly.graph_objects as go

#create a FIgure like with matplotlib
fig = go.Figure()

#add trace - plot, scatter or bar
fig.add_trace(go.Bar(
                x=[4,5,6],
                y=[2, 3, 1])
             )

#show the figure
fig.show()
```

```python
#let's try with line plot
fig = go.Figure()

#add trace - plot, scatter or bar
fig.add_trace(go.Scatter(
                x = [4,5,6],
                y = [2,3,1],
                mode = 'lines'
            ))

#show the figure
fig.show()
```

```python
#let's try with scatter plot
fig = go.Figure()

#add trace - plot, scatter or bar
fig.add_trace(go.Scatter(
                x = [4,5,6],
                y = [2,3,1],
                mode = 'markers'
            ))

#show the figure
fig.show()
```

```python
#you can add multiple trace and mix modes

fig = go.Figure()

# Create traces
fig.add_trace(go.Scatter(
    x = [1,2,3,4],
    y = [2,3,1,5],
    mode = 'lines',
    name = 'lines'
))
fig.add_trace(go.Scatter(
    x = [2,3,4,5],
    y = [2,3,4,5],
    mode = 'lines+markers',
    name = 'lines+markers'
))
fig.add_trace(go.Scatter(
    x = [1,2,3,4],
    y = [4,5,6,2],
    mode = 'markers',
    name = 'markers'
))

fig.show()
```

```python
#you can also play with layout

fig = go.Figure()

# Create traces
fig.add_trace(go.Scatter(
    x = [1,2,3,4],
    y = [2,3,1,5],
    mode = 'lines',
    name = 'lines'
))
fig.add_trace(go.Scatter(
    x = [2,3,4,5],
    y = [3,2,4,1],
    name = 'second trace',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 10,
        dash = 'dot')
))

fig.update_layout(title = 'Super cute title',
              xaxis = dict(title = 'x axis title'),
              yaxis = dict(title = 'y axis title'),
              )

fig.show()
```
